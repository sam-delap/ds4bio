## Section 1 
#### Describe the data:
This data includes fantasy football projections by a proprietary model ([Advanced Sports Logic's "The Machine"](https://advancedsportslogic.com/nfl)) and how many fantasy points the players the model projected actually scored in a given week.
#### What are the observations?
The observations are `quarterback-week` or `defense/special teams(D/ST)-week` pairs.
#### What are the features?
The features are projected fantasy points scored and actual fantasy points scored for each week of the NFL season, as well as the correlation (determined by linear regression) between the two
#### What other data or knowledge was used?
The team that each player played for, as well as a unique playerID
