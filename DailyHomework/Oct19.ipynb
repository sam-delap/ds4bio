 #### Use groupby to get the mean latitude, longitude, and elevations for sites in the E and W regions of the eco_sites data (you can get all this info in one data frame). Then choose one of those 3 features and compare these group means to one plot of your choosing (besides a histogram, so any catplot but a histogram). 
 #### Explain how the means are consistent with what you see in the catplot.

The mean (white dots in the violin plot) is slightly higher for the East (`E`) group than for the West (`W`) group, which matches the groupby results

```python
# Import packages and load data
import pandas as pd
import seaborn as sns
sns.set_theme()
sites = pd.read_table("../Data/eco_data/eco_sites.txt")

# Get the means for each group with groupby
display(sites.groupby('WorE').mean())

# Plot elevation by group in a violin plot
sns.catplot(data=sites, x='Elevation', y='WorE', kind='violin')
```
