#### How do the results compare?
Our results are similar to those in the paper. Figure 2 demonstrates that the latitude measurements are in `Â°E`, so it makes sense that sites classified as `E` in the dataset have higher latitude numbers (between 88 and 93) than those classified as `W` (between 75 and 80)

```python
# Import packages and load data
import pandas as pd
import seaborn as sns
sns.set_theme()
sites = pd.read_table("../Data/eco_data/eco_sites.txt")

# Isolate only the data we're interested in (East (latitude) and WorE)
latitude = sites.loc[:,['East', 'WorE']]

# Plot the latitude of locations in the east and locations in the west
sns.displot(data = latitude, x = 'East', col = 'WorE')
```
