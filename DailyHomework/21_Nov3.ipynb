## Make plots of fruit_length (x-axis) versus the log probability of each fruit length (y-axis) for normal models with loc = 15 but scale = 1, 2, or 5 (so either 3 plots or 3 sets of points on the same plot). 
## Describe the plots and which of the 3 models is most likely based on these plots (based on the definition of likelihood as the sum of the log probabilities).
Each plot represents the likelihood vs value of the fruit_length data. The sum of the log probabilities is probably lowest for the scale = 2 condition, since it strikes the best balance between data that is near the mean having a high log probability (value close to 0) and outlier data having a moderate log probability (-6 represents 1*10-6, which is still very unlikely).

For the scale=1 condition, some data has a log probability of -20, which is extrememly low. There is also a much wider range of values for scale=1 than for the other 2 graphs, which means that the sum of the log probabilities is likely higher

In the scale=5 condition, all of the data seems unlikely to have been generated by a loc=15 model. Even data that is close to 15 results in a log probability of -2.5, which is still unlikely.

```python
# import data and packages
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import norm

fruit_length_geno = pd.read_csv('../Data/arabmagic/fruit_length_geno.csv')
```

```python
scales = [1, 2, 5]
fruit_length = fruit_length_geno['fruit_length']
d = {0:fruit_length}
for scale in scales:
    model = norm(loc=15, scale=scale)
    probs = model.pdf(fruit_length)
    logs = np.log(probs)
    d[scale] = logs
    sns.relplot(data=d, x=0, y=scale)
```
