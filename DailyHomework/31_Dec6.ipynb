```python
# Write a linear model to model presence of birds as a function of elevation
print('log odds birds = 1/500 * elevation - 4')
```

```python
# Elevation vs probability of seeing a bird species
# Import and load data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sites = pd.read_table('../data/eco_data/eco_sites.txt')
counts = pd.read_table('../data/eco_data/species_counts.txt')

# Calculate the log odds using our predicted model
log_odds = 1/500 * sites['Elevation'] - 4

# Calculate the probability from the log odds using the formula we learned in class
prob = 1/(1 + np.exp(-1 * log_odds))

# Plot probability vs elevation
f, ax = plt.subplots(1)
ax.plot(sites['Elevation'], prob, '.')
```
