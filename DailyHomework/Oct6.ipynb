## Using the sites data, make a histogram of how common each bird species is across all 38 sites

```python
# Import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the frequency data
species = pd.read_table('../data/eco_data/species_counts.txt')

# Add together all of the columns in the dataframe
# This will give us the number of times that a bird was sighted across all 38 sites
sum_birds = species.sum(axis=0)

# Turn the resulting series back into a dataframe
sum_birds = sum_birds.to_frame()

# Rename the only column that's created (0) to be called 'Count'
sum_birds = sum_birds.rename(columns={0: 'Count'})

# Next, we'll rename the index (species of bird) to make the plotting a bit clearer
sum_birds.index = sum_birds.index.rename('Species')

# Here's some matplotlib settings that should make our plot a bit more readable
plt.figure(figsize=(100, 30))

# Create a histogram that plots how common it is to sight each bird
sns.set_theme(style="dark")
bars = sns.barplot(data=sum_birds, x=sum_birds.index, y='Count')

# Rotate the tick labels so we can kinda see everything
bars.set_xticklabels(bars.get_xticklabels(), rotation=90)

bars
```
